```python
import requests as req
from bs4 import BeautifulSoup as bs
import os, pprint, re
```

```python
url = "https://www.gutenberg.org/browse/languages/zh"

os.makedirs("project_gutenberg", exist_ok=True)
print("資料夾已建")
```

```python
gutenberg = req.get(url)
gutenberg.encoding = 'utf-8'

soup = bs(gutenberg.text, "html.parser")

links = soup.select("li.pgdbetext > a[href]")
print(f"找到 {len(links)} 本書的連結")

chinese_pattern = re.compile(r'[\u4e00-\u9fff。，、！？：「」『』（）《》—…．·]+')

for link in links:
    book_name = link.text.strip()
    href = link['href']

    if not chinese_pattern.search(book_name):
        continue
    book_url = "https://www.gutenberg.org" + href
    book_info = req.get(book_url)
    book_info.encoding = 'utf-8'
    book_soup = bs(book_info.text, "html.parser")
    txt_link = None

    for a in book_soup.select("table.files a[href]"):
        if a['href'].endswith(".txt.utf-8") and "utf-8" in a.text.lower():
            txt_link = "https://www.gutenberg.org" + a['href']
            break

    if not txt_link:
        print(f"找不到 txt 下載連結，跳過 {book_name}")
        continue

    txt_all = req.get(txt_link)
    txt_all.encoding = 'utf-8'
    text = txt_all.text

    chinese_text = ''.join(chinese_pattern.findall(text))

    filename = f"project_gutenberg/{book_name.replace('\r', '').replace('\n', '').replace('/', '').replace('\\', '').replace(':', '').replace('?', '').replace('*', '').replace('"', '').replace('<', '').replace('>', '').replace('|', '')}.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(chinese_text)

    print(f"已儲存 {filename}")


print("所有書籍處理完成！")
```
